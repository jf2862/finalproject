# Read Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import statsmodels.api as sm
```

```python
audusd=pd.read_csv('data/australia/CURRFX-AUDUSD.csv')
audusd['Date'] = pd.to_datetime(audusd['Date'])
audusd=audusd.rename(columns = {'Rate':'AUD/USD'})
```

```python
oil=pd.read_csv('data/commodity/Crude oil.csv')
oil=oil.rename(columns = {'[DateTime]':'Date',' [Close]':'Crude Oil Price'})
oil['Date'] = pd.to_datetime(oil['Date'])
```

```python
iron=pd.read_csv('data/commodity/Iron Ore.csv')
iron=iron.rename(columns = {'[DateTime]':'Date',' [Close]':'Iron Ore Price'})
iron['Date'] = pd.to_datetime(iron['Date'])
```

```python
aluminum=pd.read_csv('data/commodity/Aluminum.csv')
aluminum=aluminum.rename(columns = {'[DateTime]':'Date',' [Close]':'Aluminum Price'})
aluminum['Date'] = pd.to_datetime(aluminum['Date'])
```

```python
australia_inflation=pd.read_csv('data/australia/Australia Core Inflation Rate.csv')
australia_inflation=australia_inflation.rename(columns = {'[DateTime]':'Date',' [Close]':'Australia Inflation Rate'})
australia_inflation['Date'] = pd.to_datetime(australia_inflation['Date'])
```

```python
australia_stock=pd.read_csv('data/australia/Australia S&P ASX 200 Stock Market Index.csv')
australia_stock=australia_stock.rename(columns = {'[DateTime]':'Date',' [Close]':'Stock Market Index'})
australia_stock['Date'] = pd.to_datetime(australia_stock['Date'])
```

```python
australia_gdp_growth=pd.read_csv('data/australia/Australia GDP Growth Rate.csv')
australia_gdp_growth=australia_gdp_growth.rename(columns = {'[DateTime]':'Date',' [Close]':'GDP Growth Rate'})
australia_gdp_growth['Date'] = pd.to_datetime(australia_gdp_growth['Date'])
```

```python
australia_bot=pd.read_csv('data/australia/Australia Balance of Trade.csv')
australia_bot=australia_bot.rename(columns = {'[DateTime]':'Date',' [Close]':'Balance of Trade'})
australia_bot['Date'] = pd.to_datetime(australia_bot['Date'])
```

```python
australia_interest_rate=pd.read_csv('data/australia/Australia Interest Rate.csv')
australia_interest_rate=australia_interest_rate.rename(columns = {'[DateTime]':'Date',' [Close]':'Interest Rate'})
australia_interest_rate['Date'] = pd.to_datetime(australia_interest_rate['Date'])
```

```python
usdx=pd.read_csv('data/United States Dollar.csv')
usdx=usdx.rename(columns = {'[DateTime]':'Date',' [Close]':'USDX'})
usdx['Date'] = pd.to_datetime(usdx['Date'])
```

```python
us_federal_fund_rate=pd.read_csv('data/United States Fed Funds Rate.csv')
us_federal_fund_rate=us_federal_fund_rate.rename(columns = {'[DateTime]':'Date',' [Close]':'US Fed Funds Rate'})
us_federal_fund_rate['Date'] = pd.to_datetime(us_federal_fund_rate['Date'])
```

# Merge data

```python
aud_oil=audusd.merge(oil, how='outer')
aud_oil_al=aud_oil.merge(aluminum, how='outer')
aud_commodity=aud_oil_al.merge(iron, how='outer')
aud_commodity
```

```python
aud_commodity_inf=aud_commodity.merge(australia_inflation, how='outer')
aud_commodity_inf_grp=aud_commodity_inf.merge(australia_gdp_growth, how='outer')
aud_commodity_inf_grp_bot=aud_commodity_inf_grp.merge(australia_bot, how='outer')
aud_commodity_inf_grp_bot_interest=aud_commodity_inf_grp_bot.merge(australia_interest_rate, how='outer')
aud_commodity_inf_grp_bot_interest_usfund=aud_commodity_inf_grp_bot_interest.merge(us_federal_fund_rate, how='outer')
aud_all=aud_commodity_inf_grp_bot_interest_usfund.merge(usdx,how='outer')
```

# set date as index

```python
aud_all=aud_all.sort_values('Date', ascending=False)
```

```python
aud_all=aud_all.set_index('Date')
```

# Fill all NaNs and drop unwanted rows

```python
aud_all['Australia Inflation Rate']=aud_all['Australia Inflation Rate'].fillna(method='ffill')
aud_all['GDP Growth Rate']=aud_all['GDP Growth Rate'].fillna(method='ffill')
aud_all['Balance of Trade']=aud_all['Balance of Trade'].fillna(method='ffill')
aud_all['Interest Rate']=aud_all['Interest Rate'].fillna(method='bfill')
aud_all['US Fed Funds Rate']=aud_all['US Fed Funds Rate'].fillna(method='bfill')
```

```python
aud_all.dropna(inplace=True)
aud_all
```

# Save as csv

```python
aud_all.to_csv('australia.csv')
```
