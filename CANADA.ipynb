```python
canada_interest_rate=pd.read_csv('/Users/tongshan/Desktop/final/data/canada/Canada Interest Rate.csv')
canada_interest_rate=canada_interest_rate.rename(columns = {'[DateTime]':'Date',' [Close]':'Interest Rate'})
canada_interest_rate['Date'] = pd.to_datetime(canada_interest_rate['Date'])
```

```python
canada_m2=pd.read_csv('/Users/tongshan/Desktop/final/data/canada/Canada Money Supply M2.csv')
canada_m2=canada_m2.rename(columns = {'[DateTime]':'Date',' [Close]':'Money Supply'})
canada_m2['Date'] = pd.to_datetime(canada_m2['Date'])
```

```python
usdx=pd.read_csv('/Users/tongshan/Desktop/final/data/United States Dollar.csv')
usdx=usdx.rename(columns = {'[DateTime]':'Date',' [Close]':'USDX'})
usdx['Date'] = pd.to_datetime(usdx['Date'])
```

```python
us_federal_fund_rate=pd.read_csv('/Users/tongshan/Desktop/final/data/United States Fed Funds Rate.csv')
us_federal_fund_rate=us_federal_fund_rate.rename(columns = {'[DateTime]':'Date',' [Close]':'US Fed Funds Rate'})
us_federal_fund_rate['Date'] = pd.to_datetime(us_federal_fund_rate['Date'])
```

# Merge all data

```python
cad_oil=cadusd.merge(oil, how='outer')
cad_oil_al=cad_oil.merge(aluminum, how='outer')
cad_commodity=cad_oil_al.merge(iron, how='outer')
cad_commodity
```

```python
cad_commodity_inf=cad_commodity.merge(canada_inflation, how='outer')
cad_commodity_inf_grp=cad_commodity_inf.merge(canada_gdp_growth, how='outer')
cad_commodity_inf_grp_bot=cad_commodity_inf_grp.merge(canada_bot, how='outer')
cad_commodity_inf_grp_bot_interest=cad_commodity_inf_grp_bot.merge(canada_interest_rate, how='outer')
cad_commodity_inf_grp_bot_interest_usfund=cad_commodity_inf_grp_bot_interest.merge(us_federal_fund_rate, how='outer')
cad_commodity_inf_grp_bot_interest_usfund_m2=cad_commodity_inf_grp_bot_interest_usfund.merge(canada_m2, how='outer')
cad_all=cad_commodity_inf_grp_bot_interest_usfund_m2.merge(usdx,how='outer')
```

### Set date as index

```python
cad_all=cad_all.sort_values('Date', ascending=False)
```

```python
cad_all=cad_all.set_index('Date')
```

### Fill all NaNs and drop unwanted rows

```python
cad_all['Canadian Inflation Rate']=cad_all['Canadian Inflation Rate'].fillna(method='ffill')
cad_all['GDP Growth Rate']=cad_all['GDP Growth Rate'].fillna(method='ffill')
cad_all['Money Supply']=cad_all['Money Supply'].fillna(method='ffill')
cad_all['Balance of Trade']=cad_all['Balance of Trade'].fillna(method='ffill')
cad_all['Interest Rate']=cad_all['Interest Rate'].fillna(method='bfill')
cad_all['US Fed Funds Rate']=cad_all['US Fed Funds Rate'].fillna(method='bfill')
```

```python
cad_all.dropna(inplace=True)
```

```python
cad_all
```

### Save as csv

```python
cad_all.to_csv('canada.csv')
```
