```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import statsmodels.api as sm
```

# Read data

```python
jpyusd=pd.read_csv('data/Japan/CURRFX-JPYUSD.csv')
jpyusd['Date'] = pd.to_datetime(jpyusd['Date'])
jpyusd=jpyusd.rename(columns = {'Rate':'JPY/USD'})
```

```python
oil=pd.read_csv('data/commodity/Crude oil.csv')
oil=oil.rename(columns = {'[DateTime]':'Date',' [Close]':'Crude Oil Price'})
oil['Date'] = pd.to_datetime(oil['Date'])
```

```python
iron=pd.read_csv('data/commodity/Iron Ore.csv')
iron=iron.rename(columns = {'[DateTime]':'Date',' [Close]':'Iron Ore Price'})
iron['Date'] = pd.to_datetime(iron['Date'])
```

```python
aluminum=pd.read_csv('data/commodity/Aluminum.csv')
aluminum=aluminum.rename(columns = {'[DateTime]':'Date',' [Close]':'Aluminum Price'})
aluminum['Date'] = pd.to_datetime(aluminum['Date'])
```

```python
japan_inflation=pd.read_csv('data/japan/Japan Core Inflation Rate.csv')
japan_inflation=japan_inflation.rename(columns = {'[DateTime]':'Date',' [Close]':'Japan Inflation Rate'})
japan_inflation['Date'] = pd.to_datetime(japan_inflation['Date'])
```

```python
japan_stock=pd.read_csv('data/japan/Japan NIKKEI 225 Stock Market Index.csv')
japan_stock=japan_stock.rename(columns = {'[DateTime]':'Date',' [Close]':'Stock Market Index'})
japan_stock['Date'] = pd.to_datetime(japan_stock['Date'])
```

```python
japan_gdp_growth=pd.read_csv('data/japan/Japan GDP Growth Rate.csv')
japan_gdp_growth=japan_gdp_growth.rename(columns = {'[DateTime]':'Date',' [Close]':'GDP Growth Rate'})
japan_gdp_growth['Date'] = pd.to_datetime(japan_gdp_growth['Date'])
```

```python
japan_bot=pd.read_csv('data/japan/Japan Balance of Trade.csv')
japan_bot=japan_bot.rename(columns = {'[DateTime]':'Date',' [Close]':'Balance of Trade'})
japan_bot['Date'] = pd.to_datetime(japan_bot['Date'])
```

```python
japan_interest_rate=pd.read_csv('data/japan/Japan Interest Rate.csv')
japan_interest_rate=japan_interest_rate.rename(columns = {'[DateTime]':'Date',' [Close]':'Interest Rate'})
japan_interest_rate['Date'] = pd.to_datetime(japan_interest_rate['Date'])
```

```python
usdx=pd.read_csv('/Users/tongshan/Desktop/final/data/United States Dollar.csv')
usdx=usdx.rename(columns = {'[DateTime]':'Date',' [Close]':'USDX'})
usdx['Date'] = pd.to_datetime(usdx['Date'])
```

```python
us_federal_fund_rate=pd.read_csv('/Users/tongshan/Desktop/final/data/United States Fed Funds Rate.csv')
us_federal_fund_rate=us_federal_fund_rate.rename(columns = {'[DateTime]':'Date',' [Close]':'US Fed Funds Rate'})
us_federal_fund_rate['Date'] = pd.to_datetime(us_federal_fund_rate['Date'])
```

# Merge all data

```python
jap_oil=jpyusd.merge(oil, how='outer')
jap_oil_al=jap_oil.merge(aluminum, how='outer')
jap_commodity=jap_oil_al.merge(iron, how='outer')
jap_commodity
```

```python
jap_commodity_inf=jap_commodity.merge(japan_inflation, how='outer')
jap_commodity_inf_grp=jap_commodity_inf.merge(japan_gdp_growth, how='outer')
jap_commodity_inf_grp_bot=jap_commodity_inf_grp.merge(japan_bot, how='outer')
jap_commodity_inf_grp_bot_interest=jap_commodity_inf_grp_bot.merge(japan_interest_rate, how='outer')
jap_commodity_inf_grp_bot_interest_usfund=jap_commodity_inf_grp_bot_interest.merge(us_federal_fund_rate, how='outer')

jap_all=jap_commodity_inf_grp_bot_interest_usfund.merge(usdx,how='outer')
```

### Set date as index

```python
jap_all=jap_all.sort_values('Date', ascending=False)
```

```python
jap_all=jap_all.set_index('Date')
```

### Fill all NaNs and drop unwanted rows

```python
jap_all['Japan Inflation Rate']=jap_all['Japan Inflation Rate'].fillna(method='ffill')
jap_all['GDP Growth Rate']=jap_all['GDP Growth Rate'].fillna(method='ffill')
jap_all['Balance of Trade']=jap_all['Balance of Trade'].fillna(method='ffill')
jap_all['Interest Rate']=jap_all['Interest Rate'].fillna(method='bfill')
jap_all['US Fed Funds Rate']=jap_all['US Fed Funds Rate'].fillna(method='bfill')
```

```python
jap_all.dropna(inplace=True)
```

```python
jap_all
```

### Save as csv

```python
jap_all.to_csv('japan.csv')
```
